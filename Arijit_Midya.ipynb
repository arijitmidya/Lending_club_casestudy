1. Every visual representation is a comparison between loan_status and a numerical variable
2. Boxplot is used for the analysis
3. The boxplot highlights the distribution(min , Lower whiskers , Quartile 1 , median , Quartile 3 , upper whiskers , max )  of numerical variable with respect to different loan_status
4. We can infer the numerical variable is not significaant in case the major distribution of data is similar for "Fully paid" and "Charged off" as loan status
5. We can infer the  numerical variable is significaant in case the major distribution of data is different for "Fully paid" and "Charged off" as loan status and recommend the numerical variable to be considered going forward while approving a loan
